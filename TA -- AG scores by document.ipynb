```python
# this would become Fig 5 in the assignment

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

data = pd.read_csv() # top AG scores saved as csv

data.sort_values('AG Score', ascending=False, inplace=True)

colors = []
for file in data['File']:
    if 'LeeKuanYew' in file:
        colors.append('#ff0000')  # red for LKY
    elif 'LeeHsienLoong' in file:
        colors.append('#ffa500')  # orange for LHL
    elif 'SRajaratnam' in file:
        colors.append('#0000ff')  # blue for Raja
    else:
        colors.append('#d3d3d3')  # grey for others

plt.figure(figsize=(15, 10))
barlist = plt.bar(data['File'], data['AG Score'], color=colors)

plt.xticks(rotation=90, fontsize=8)

# labels
plt.xlabel('Document', fontsize=12)
plt.ylabel('AG Score', fontsize=12)

# legend
legend_elements = [Patch(facecolor='#ff0000', label='Lee Kuan Yew'),
                   Patch(facecolor='#ffa500', label='Lee Hsien Loong'),
                   Patch(facecolor='#0000ff', label='S Rajaratnam'),
                   Patch(facecolor='#d3d3d3', label='Others')]

plt.legend(handles=legend_elements, loc='upper right', title='Legend')

plt.tight_layout()

plt.show()
```
